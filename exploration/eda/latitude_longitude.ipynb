# LATITUDE & LONGITUDE

```python
import pandas as pd

# import data
path = '../../data/raw/UJIndoorLoc/'
data = pd.read_csv(path + 'trainingData.csv')
valid = pd.read_csv(path + 'validationData.csv')

# define wap names
wap_names = data.iloc[:,:520].columns.tolist()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, train_size=0.8, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor 
```

```python
knn_lat = KNeighborsRegressor()
```

```python
knn_lat.fit(train[wap_names], train['LATITUDE'])
```

```python
preds_lat = knn_lat.predict(test[wap_names])
```

```python
from sklearn import metrics

def regression_metrics(y_true, y_preds):
    rmse = metrics.mean_squared_error(y_true, y_preds)
    mea = metrics.mean_absolute_error(y_true, y_preds)
    eva = metrics.explained_variance_score(y_true, y_preds)
    r2 = metrics.r2_score(y_true, y_preds)
    
    print('RMSE: ', rmse)
    print('MEA: ', mea)
    print('Explained variance score: ', eva)
    print('R^2: ', r2)
```

```python
regression_metrics(y_true=test['LATITUDE'], y_preds=preds_lat)
```

```python
knn_lon = KNeighborsRegressor()
```

```python
knn_lon.fit(train[wap_names], train['LONGITUDE'])
```

```python
preds_lon = knn_lon.predict(test[wap_names])
```

```python
regression_metrics(y_true=test['LONGITUDE'], y_preds=preds_lon)
```

```python
test['LONGITUDE_pred'] = preds_lon
test['LATITUDE_pred'] = preds_lat
```
