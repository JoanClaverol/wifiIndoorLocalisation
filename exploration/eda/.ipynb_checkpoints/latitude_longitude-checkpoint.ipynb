# LATITUDE & LONGITUDE

```python
import pandas as pd

# import data
path = '../../data/raw/UJIndoorLoc/'
data = pd.read_csv(path + 'trainingData.csv')
valid = pd.read_csv(path + 'validationData.csv')

# define wap names
wap_names = data.iloc[:,:520].columns.tolist()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(data, train_size=0.8, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsRegressor 
```

```python
knn_lat = KNeighborsRegressor()
```

```python
knn_lat.fit(train[wap_names], train['LATITUDE'])
```

```python
preds_lat_test = knn_lat.predict(test[wap_names])
```

```python
from sklearn import metrics

def regression_metrics(y_true, y_preds):
    rmse = metrics.mean_squared_error(y_true, y_preds)
    mea = metrics.mean_absolute_error(y_true, y_preds)
    eva = metrics.explained_variance_score(y_true, y_preds)
    r2 = metrics.r2_score(y_true, y_preds)
    
    print('RMSE: ', rmse)
    print('MEA: ', mea)
    print('Explained variance score: ', eva)
    print('R^2: ', r2)
```

```python
regression_metrics(y_true=test['LATITUDE'], y_preds=preds_lat)
```

```python
preds_lat_valid = knn_lat.predict(valid[wap_names])
```

```python
knn_lon = KNeighborsRegressor()
```

```python
knn_lon.fit(train[wap_names], train['LONGITUDE'])
```

```python
preds_lon = knn_lon.predict(test[wap_names])
```

```python
regression_metrics(y_true=test['LONGITUDE'], y_preds=preds_lon)
```

```python
def plot_results(df, lat_preds, lon_preds):
    # libraries
    from pandas import DataFrame
    import seaborn as sns
    import matplotlib.pyplot as plt
    # create df
    temp_data = DataFrame({
        'LON': df['LONGITUDE'],
        'p_LON': lon_preds, 
        'LAT': df['LATITUDE'],
        'p_LAT': lat_preds,
        'FLOOR': df['FLOOR']
    })
    # plot the results
    # facets
    g = sns.FacetGrid(temp_data, col='FLOOR', col_wrap=3, aspect=1, height=5)
    # scatter
    g.map(sns.scatterplot, 'p_LON', 'p_LAT', alpha=0.3, color='r')
    g.map(sns.scatterplot, 'LON', 'LAT', color='b')
    # style 
    sns.set_style("white")
    sns.despine()
    # size
    plt.rcParams['figure.figsize'] = [40, 20]
    plt.show()
```

```python
# define df
temp_df=test

# LATITUDE
knn_lat = KNeighborsRegressor()

knn_lat.fit(train[wap_names], train['LATITUDE'])
preds_lat = knn_lat.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LATITUDE'], y_preds=preds_lat)


# LONGITUDE
knn_lon = KNeighborsRegressor()

knn_lon.fit(train[wap_names], train['LONGITUDE'])
preds_lon = knn_lon.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LONGITUDE'], y_preds=preds_lon)

plot_results(df=test, lat_preds=preds_lat, lon_preds=preds_lon)
```

```python
# define df
temp_df=valid

# LATITUDE
knn_lat = KNeighborsRegressor()

knn_lat.fit(train[wap_names], train['LATITUDE'])
preds_lat = knn_lat.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LATITUDE'], y_preds=preds_lat)


# LONGITUDE
knn_lon = KNeighborsRegressor()

knn_lon.fit(train[wap_names], train['LONGITUDE'])
preds_lon = knn_lon.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LONGITUDE'], y_preds=preds_lon)

# PLOT
plot_results(df=valid, lat_preds=preds_lat, lon_preds=preds_lon)
```

## Normalization by row

```python
# 3. Normalize the waps by row by min and max
norm_df = data.copy()
norm_valid = valid.copy()

norm_df[wap_names] = norm_df[wap_names].apply(lambda x: (x - x.min())/(x.max() - x.min()), axis='columns', result_type='expand')
#norm_df['id'] = str(norm_df.index + 1)
norm_df_clean = norm_df.dropna() # missing values are created in 76 rows, as they have 0 variance (further exploration)
norm_valid[wap_names] = norm_valid[wap_names].apply(lambda x: (x - x.min())/(x.max() - x.min()), axis='columns', result_type='expand')
```

```python
train, test = train_test_split(norm_df_clean, train_size=0.8, random_state=42)
```

```python
# define df
temp_df=test

# LATITUDE
knn_lat = KNeighborsRegressor()

knn_lat.fit(train[wap_names], train['LATITUDE'])
preds_lat = knn_lat.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LATITUDE'], y_preds=preds_lat)


# LONGITUDE
knn_lon = KNeighborsRegressor()

knn_lon.fit(train[wap_names], train['LONGITUDE'])
preds_lon = knn_lon.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LONGITUDE'], y_preds=preds_lon)

# PLOT
plot_results(df=temp_df, lat_preds=preds_lat, lon_preds=preds_lon)
```

```python
# define df
temp_df=norm_valid

# LATITUDE
knn_lat = KNeighborsRegressor()

knn_lat.fit(train[wap_names], train['LATITUDE'])
preds_lat = knn_lat.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LATITUDE'], y_preds=preds_lat)


# LONGITUDE
knn_lon = KNeighborsRegressor()

knn_lon.fit(train[wap_names], train['LONGITUDE'])
preds_lon = knn_lon.predict(temp_df[wap_names])
regression_metrics(y_true=temp_df['LONGITUDE'], y_preds=preds_lon)

# PLOT
plot_results(df=temp_df, lat_preds=preds_lat, lon_preds=preds_lon)
```
